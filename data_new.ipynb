```python
import numpy as np
import pandas as pd

# Define the data science disciplines and abbreviations
disciplines = [
    "Data Scientist",
    "Machine Learning Engineer",
    "Data Engineer",
    "Research Scientist",
    "Data Analyst",
    "Research Engineer",
    "Analytics Engineer",
    "Applied Scientist",
    "AI Developer",
    "Business Intelligence Analyst",
    "Data Architect"
]

abbreviations = [
    "DS",
    "MLE",
    "DE",
    "RS",
    "DA",
    "RE",
    "AE",
    "AS",
    "AID",
    "BIA",
    "DA"
]

# Define the annual pay packages for each role (sample values in USD)
annual_pay = [
    120000,  # Data Scientist
    115000,  # Machine Learning Engineer
    110000,  # Data Engineer
    105000,  # Research Scientist
    90000,   # Data Analyst
    95000,   # Research Engineer
    100000,  # Analytics Engineer
    108000,  # Applied Scientist
    125000,  # AI Developer
    95000,   # Business Intelligence Analyst
    130000   # Data Architect
]

# Generate salary data for each of the 20 years with increases every 4 years
np.random.seed(42)  # Set seed for reproducibility
years = 20

# Create a dictionary to hold salary data for each year
salary_data = {}
for i in range(years):
    year = 2004 + i
    # Every 4 years, increase the salary based on the initial pay, ensuring random distinct values
    if i % 4 == 0 and i > 0:
        annual_pay = [pay + np.random.randint(1000, 1500) for pay in annual_pay]  # Increment salaries
    salary_data[f"Year_{year}"] = np.random.randint(low=np.array(annual_pay) - 15000,
                                                     high=np.array(annual_pay) + 30000)

# Create DataFrame
data = {
    "Data Science Discipline": disciplines,
    "Abbreviation": abbreviations,
    "Annual Pay Package (USD)": annual_pay,
}

# Update with the generated salary data for the years
data.update(salary_data)

# Convert to a DataFrame for better visualization
df1 = pd.DataFrame.from_dict(data)

# Display the DataFrame
print(df1)
```

```python
df1 = pd.DataFrame(data)
```

```python
# Optional: Save to CSV
df1.to_csv("ds1_salaries.csv", index=False)
```
